hue_order kategorik veri içerisinde yoğunlaşmak istediğimiz grupları seçmemize yarar. Yukarıdaki grafik Lunch ve Dinner kategorik verilerini hue parametresi ile günlere ayırarak verdi. hue_order ise bunlar arasından istediğini seçme imkanı veriyor. Örneğin aşağıda Lunch ve Dinner için sadece Thur ve Sun daylarine odaklanalım.

```python
sns.countplot(x="time",hue="day",hue_order = ["Thur","Sun"], data=tips)
```

Color Parametresi için tanımlı renkler girilebilir.

```python
sns.countplot(x="day",data=tips, color = "r")
```

pallette seaborn tarafından gruplandırılmış renkler bütünüdür. Sıra ile kaç tane kategorik değişken varsa rengi atar. Genelde 10 renk olur. Boxların kenarlarına çizgi vermek için lw ile genişlik ayarlanır, ec ile renk atanır ve hatch ile içine desen için değer verilir.

```python
sns.countplot(x="day",data=tips, palette="CMRmap", lw =5, ec="black",hatch="o")
```

```python
sns.countplot(x="day",data=tips, palette = "twilight")
```

the whiskers have a length of 1.5*IQR. upper ve lower değerleri bulmak için IQR ı 1.5 ile çarpıyorduk bunlar kedinin bıyığı gibi düşünün whiskerslar. İşte bu 1.5 değeri whis değeri oluyor. Bunun artırılması ile outlier bazı değerler verinin içinde kalabiliyor yada dışında olabiliyor.

```python
sns.boxplot(x="total_bill",y="day",data = tips,palette = "copper",whis=2.5)
```

scale{“area”, “count”, “width”}, optional
The method used to scale the width of each violin. If area, each violin will have the same area. If count, the width of the violins will be scaled by the number of observations in that bin. If width, each violin will have the same width.

```python
sns.violinplot(x = "day", y = "tip",hue="sex", data = tips, split=True,scale="count")
```
